```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
CA_file = "Resources/CAvideos.csv"
CA_df = pd.read_csv(CA_file)
CA_df.head()
```

```python
MX_file = "Resources/MXvideos.csv"
MX_df = pd.read_csv(MX_file, encoding="ISO-8859-1")
MX_df.head()
```

```python
US_file = "Resources/USvideos.csv"
US_df = pd.read_csv(US_file)
US_df.head()
```

### Transform Canada DataFrame

```python
# Create a filtered DataFrame from specific columns
CA_cols = ["video_id", "trending_date", "title", "category_id", "publish_time",
           "views", "likes", "dislikes", "comment_count"]
CA_transformed = CA_df[CA_cols].copy()

# Add a column to specify country
CA_transformed.insert(5, "country", "Canada")

# Create a dict to convert category_id to category name
categories = {
    1:"Film & Animation", 2:"Autos & Vehicles", 10:"Music", 15:"Pets & Animals",
    17:"Sports", 18:"Short Movies", 19:"Travel & Events", 20:"Gaming", 21:"Videoblogging",
    22:"People & Blogs", 23:"Comedy", 24:"Entertainment", 25:"News & Politics",
    26:"Howto & Style", 27:"Education", 28:"Science & Technology", 29:"Nonprofits & Activism",
    30:"Movies", 31:"Anime/Animation", 32:"Action/Adventure", 33:"Classics", 34:"Comedy",
    35:"Documentary", 36:"Drama", 37:"Family", 38:"Foreign", 39:"Horror", 40:"Sci-Fi/Fantasy",
    41:"Thriller", 42:"Shorts", 43:"Shows", 44:"Trailers"}

# Convert category_id to category
CA_transformed["category_id"].replace(categories, inplace=True)

# Change "category_id" column name to "category"
CA_transformed = CA_transformed.rename(columns={"category_id":"category"})

# Clean the data by dropping duplicates and setting the index
CA_transformed.drop_duplicates("video_id", inplace=True)
CA_transformed.set_index("video_id", inplace=True)
#CA_transformed = CA_transformed.sort_values(by=["views", "likes"], ascending=False)

CA_transformed.head()
```
