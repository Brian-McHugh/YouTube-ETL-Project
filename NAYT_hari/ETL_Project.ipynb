```python
data1 = pd.read_sql("select title, max(views) as max_views from na_titles group by title order by max_views desc limit 10;", conn)
```

```python
data1.head()
```

```python
title = data1["title"]
views = data1["max_views"]
```

```python
plt.barh(title, views, color='r', alpha=0.5, align="center")
plt.xlabel('title')
plt.title('Youtube top 10 in NA')
plt.savefig("YT_NA_TOP_10.jpeg", bbox_inches = "tight")
plt.show()
```
